Step 2: Load Datasets

```python
import pandas as pd

# Load the datasets (ALL rows will be displayed)
transactions_df = pd.read_csv("transactions_data.csv")
customers_df = pd.read_csv("customers_data.csv")
products_df = pd.read_csv("products_data.csv")

# Display full datasets
print("\nTransactions Data:")
display(transactions_df)  # Shows ALL rows

print("\nCustomers Data:")
display(customers_df)  # Shows ALL rows

print("\nProducts Data:")
display(products_df)  # Shows ALL rows

# Confirm dataset size
print("\nTransactions Data Shape:", transactions_df.shape)
print("Customers Data Shape:", customers_df.shape)
print("Products Data Shape:", products_df.shape)
```

 Step 3: Check for Missing Values and Duplicates

```python
# Check for missing values
print("\nMissing Values in Transactions Data:")
print(transactions_df.isnull().sum())

print("\nMissing Values in Customers Data:")
print(customers_df.isnull().sum())

print("\nMissing Values in Products Data:")
print(products_df.isnull().sum())

# Check for duplicate rows
print("\nDuplicate Rows in Transactions:", transactions_df.duplicated().sum())
print("Duplicate Rows in Customers:", customers_df.duplicated().sum())
print("Duplicate Rows in Products:", products_df.duplicated().sum())
```

 Step 4: Handle Missing Values and Fix Data Formats

```python
# Convert Transaction_Date to datetime
transactions_df['Transaction_Date'] = pd.to_datetime(transactions_df['Transaction_Date'], errors='coerce')

# Convert Product_Price to numeric (removing currency symbols)
for df in [transactions_df, customers_df, products_df]:
    if 'Product_Price' in df.columns:
        df['Product_Price'] = pd.to_numeric(df['Product_Price'].astype(str).str.replace(r'[^\d.]', '', regex=True), errors='coerce')

# Fill missing values
transactions_df.fillna({'Quantity': 0, 'Total_Cost': transactions_df['Total_Cost'].median()}, inplace=True)
customers_df.fillna({'Company_Profit': customers_df['Company_Profit'].median(), 'Company_ID': -1}, inplace=True)
products_df.fillna({'Product_ID': -1}, inplace=True)

# Display cleaned datasets
print("\nCleaned Transactions Data:")
display(transactions_df)

print("\nCleaned Customers Data:")
display(customers_df)

print("\nCleaned Products Data:")
display(products_df)
```

Step 5: Merge Datasets

```python
# Merge Transactions with Products using Product_ID
merged_df = transactions_df.merge(products_df, on="Product_ID", how="left")

# Merge the result with Customers using Company_ID
merged_df = merged_df.merge(customers_df, on="Company_ID", how="left")

# Drop unnecessary columns
merged_df.drop(columns=['Transaction_ID', 'Transaction_Date', 'Address'], errors='ignore', inplace=True)

# Display the merged dataset (ALL ROWS)
print("\nMerged Dataset:")
display(merged_df)

# Confirm merged dataset size
print("\nMerged Dataset Shape:", merged_df.shape)
```

```python
# Identify correct Product_Price column
if 'product_price_x' in merged_df.columns:
    merged_df.rename(columns={'product_price_x': 'product_price'}, inplace=True)
elif 'product_price_y' in merged_df.columns:
    merged_df.rename(columns={'product_price_y': 'product_price'}, inplace=True)

# Confirm column names again
print("\nUpdated Column Names After Fixing:", merged_df.columns.tolist())
```

 Step 6: Feature Engineering

```python
from sklearn.preprocessing import StandardScaler

# Standardize correct column names
columns_to_scale = ['product_price', 'total_cost', 'company_profit']

# Ensure columns exist before scaling
columns_to_scale = [col for col in columns_to_scale if col in merged_df.columns]

# Apply scaling
scaler = StandardScaler()
merged_df[columns_to_scale] = scaler.fit_transform(merged_df[columns_to_scale])

# Display final dataset
print("\nFeature Engineered Data:")
display(merged_df)
```

Step 7: Train Predictive Models
